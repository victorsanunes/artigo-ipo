```python
from sklearn.datasets import make_classification
import itertools
```

```python
n_features = [2, 5, 10]
error_rate = [0.01, 0.1, 0.5]
N = 1000

for product in itertools.product(n_features, error_rate):
    x, y = make_classification(n_samples = N, 
                                n_features = product[0],
                                n_informative = product[0], 
                                n_redundant = 0, 
                                n_repeated = 0, 
                                n_classes = 2,
                                flip_y = product[1],
                                scale = 0.1,
                                random_state = 42)
    x_df = pd.DataFrame(x, columns = ['x'+str(i) for i in range(product[0])])
    y_df = pd.DataFrame(y, columns = ['target'])
    dataset = pd.concat([x_df, y_df], axis = 1)
    filename = str(product[0]) + '_features_' + str(product[1]) + '_error' + '.csv' 
    dataset.to_csv(f'datasets/sklearn-datasets/{filename}')
```
