```python
import os 
import pandas as pd

synthetic_datasets = [f for f in os.listdir('datasets/sklearn-datasets', ) if f.endswith('.csv')]
synthetic_datasets
```

```python
datasets_dictionary = dict()
for filename in synthetic_datasets:
    try:
        df = pd.read_csv(f'datasets/sklearn-datasets/{filename}')
        target_columns = [col for col in df.columns if col.startswith('target')]
        datasets_dictionary[filename] = (df, target_columns)
    except:
        pass
```

```python
from sklearn.model_selection import train_test_split

splitted_dataset = dict()

for filename in datasets_dictionary:
    raw_df = datasets_dictionary[filename][0]
    target_columns = datasets_dictionary[filename][1]
    
    X = raw_df.drop(columns = target_columns)
    y = raw_df[target_columns]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    splitted_dataset[filename] = X_train, X_test, y_train, y_test
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.pipeline import Pipeline


def train_model(model, X_train, y_train, params):
    cv = RepeatedKFold(n_splits = 5, n_repeats = 1, random_state=1)

    gb_tuned = GridSearchCV(estimator = model, 
                            param_grid = params,
                            cv = cv, verbose = 1)
    gb_tuned.fit(X_train, y_train)
    return gb_tuned.best_estimator_
```

```python
%%time

pipeline_steps = steps=[
                        ('model', GradientBoostingClassifier())
                       ]
classification_pipeline = Pipeline(pipeline_steps)

params = { 'model__n_estimators': [16, 32], 'model__learning_rate': [0.1, 0.8, 1.0] }

models_trained = dict()
for filename in splitted_dataset:
    X_train = splitted_dataset[filename][0].values
    y_train = splitted_dataset[filename][2].values
    
    models_trained[filename] = train_model(classification_pipeline, X_train, y_train, params)
```

```python

classification_output = dict()
for filename in datasets_dictionary:
    model = models_trained[filename]
    raw_df = datasets_dictionary[filename][0]
    target_columns = datasets_dictionary[filename][1]
    
    X = raw_df.drop(columns = target_columns)
    classification_output[filename] = model.predict_proba(X)
    print(classification_output[filename].shape)
```

```python
for filename in classification_output:
    df = pd.DataFrame(classification_output[filename][:,0], columns=["algo_gradient_boosting"])
    df.to_csv(f"datasets/sklearn-datasets/performance/{filename}", index = False)
```
